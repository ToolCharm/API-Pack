```python
from azureml.core import Workspace, Dataset

subscription_id = '76ea09d7-5d6b-45bb-b1c9-f504ad119f67'
resource_group = 'beta-resources'
workspace_name = 'open-source-beta'

workspace = Workspace(subscription_id, resource_group, workspace_name)

dataset = Dataset.get_by_id(workspace, 'c5833ea9-972b-4778-8c86-2117bfcde1b8')
dataset.download(target_path='.', overwrite=False)
```

```python
import os
import pandas as pd

# Initialize an empty list to store individual DataFrames
dataframes = []

# Loop through all files in the current directory
for filename in os.listdir('.'):
    # Check if the file is a CSV file
    if filename.endswith('.csv') and "shard" in filename:
        print(f"Processing file: {filename}")
        # number = int(filename.split("shard_")[1].split(".csv")[0])
        # if number > 132:
        #     continue
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(filename)
        # Append the DataFrame to the list
        dataframes.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
combined_dataframe = pd.concat(dataframes, ignore_index=True)
combined_dataframe.to_csv("aa_rest_base_generated_instructions.csv", index=False)
```
